```python
# this works for page 2
# next we do a loop for more pages

web_address='https://www.presidency.ucsb.edu/documents/app-categories/presidential/spoken-addresses-and-remarks?page=1'
url = 'https://www.presidency.ucsb.edu'
web_page = urllib.request.urlopen(web_address) # open the web page
soup = BeautifulSoup(web_page.read()) 
all_titles = soup.find_all('div',{'class': 'field-title'})

with open('hw2_test.csv', 'w') as f: # set up with the writer
  w = csv.DictWriter(f, fieldnames = ("date", "title","text", "footnote", "citation",)) # define column names
  w.writeheader() # write the header
  speech = {}
  for i in range(0,10): 
    try:
      inner_page_url = url + all_titles[i].a['href'] 
      inner_page = urllib.request.urlopen(inner_page_url) # open the address page
      inner_soup = BeautifulSoup(inner_page.read()) # soup the page
      speech['date'] = inner_soup.find('span',{'class':'date-display-single'}).text
      speech['title'] = inner_soup.find('h1').text
      speech['text'] = inner_soup.find('div',{'class':'field-docs-content'}).text
      speech['footnote'] = inner_soup.find('div',{'class':'field-docs-footnote'}).text
      speech['citation'] = inner_soup.find('p',{'class':'ucsbapp_citation'}).text
      
    except:
      pass
      
    w.writerow(speech) # write the row 
    time.sleep(random.uniform(1, 5)) # be polite, sleep!
```
